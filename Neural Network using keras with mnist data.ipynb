```python
import numpy as np
import matplotlib.pyplot as plt

from keras.datasets import mnist
from keras.models import Model
from keras.callbacks import TensorBoard
from keras.layers import Input, Dense, Reshape, Dropout
from keras.optimizers import SGD, Adam

from keras.models import model_from_json
```

```python
lr = 0.01
wd = 0.01

batch_size = 128
num_epochs = 5
```

```python
def create_data():
    (x_train, y_train), (x_test, y_test) = mnist.load_data()

    x_train = x_train.reshape(60000,-1)
    y_train = y_train.reshape(60000,-1)

    x_test = x_test.reshape(10000,-1)
    y_test = y_test.reshape(10000,-1)

    return x_train, y_train, x_test, y_test
```

```python
def buildModel(x_train, y_train, x_test, y_test):

    input_img = Input(shape=(784,))

    dense_1 = Dense(512,activation='sigmoid')(input_img)

    dense_2 = Dense(256, activation = 'sigmoid')(dense_1)

    dense_3 = Dense(256, activation = 'sigmoid')(dense_2)

    dense_4 = Dense(256, activation = 'sigmoid')(dense_3)

    output = Dense(1, activation='sigmoid')(dense_4)

    model = Model(input_img, output)

    # optimizer
    sgd = SGD(lr=lr, decay=1e-3, momentum=0.9, nesterov=True, clipvalue=5.0)

    model.compile(loss='mean_squared_error', optimizer=sgd,metrics=['accuracy'])

    print (model.summary())
    
    model.fit(x_train, y_train, 
    nb_epoch=num_epochs, 
    batch_size=batch_size, 
    shuffle=True, 
    validation_data=(x_test, y_test))

    return model
```

```python
def main():
    # getting the data
    x_train, y_train, x_test, y_test = create_data()

    model = buildModel(x_train, y_train, x_test, y_test)

    # decoded_imgs = autoencoder.predict(x_test)

    pass
```

```python
if __name__ == '__main__':
    main()
```
